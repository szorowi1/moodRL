## Section 2: Model fitting

```python
import os, pystan
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import _pickle as cPickle
from pandas import DataFrame, read_csv
from scripts.diagnostics import check_div, check_energy, check_treedepth
from scripts.plotting import plot_subject
sns.set_style('white')
sns.set_context('notebook', font_scale=1.75)
%matplotlib inline
%load_ext jupyternotify

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Define parameters.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Select model.
model_name = False

## Sampling parameters.
samples = 1250
warmup = 1000
chains = 4
thin = 1
n_jobs = 4

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Load and prepare data.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Load and prepare behavior data.
data = read_csv('data/moodRL_data.csv')
data = data[data.Block < 4]
data = data.fillna(-1)

## Load and prepare ratings data.
ratings = read_csv('data/moodRL_ratings.csv')
ratings = ratings[ratings.Variable=='Mood']

## Load and prepare metadata.
metadata = read_csv('data/moodRL_metadata.csv')

## Define metadata.
N = data.Datetime.unique().size
B = data.Block.max()
T = data.Trial.max()

## Extract and prepare stimulus presentation data.
X = data[['M1','M2']].values.reshape(N,B,T,2)

## Extract and prepare choice data. 
## Missing values coded as -1.
Y = data.Choice.values.reshape(N,B,T,1)
Y = np.where( np.any(np.equal(X, Y), axis=-1), np.argmax(np.equal(X, Y), axis=-1) + 1, -1 )

## Extract and prepare reward data. All rewards 
## scaled between [0,1].
R = data.Outcome.values.reshape(N,B,T)
R = np.where(R > 0, 1, 0)

## Extract and prepare mood data.
M = ratings.loc[ratings.Trial>0, 'Rating'].values.reshape(N,B,3) / 4

m2 = ratings.loc[np.logical_and(ratings.Block==2, ratings.Trial==0),'Rating'].values / 4
m2 = np.where(m2==-1, -0.99, np.where(m2==1, 0.99, m2))

## Organize data dictionary.
dd = dict(N=N, B=B, T=T, X=X, Y=Y, R=R, M=M, m2=m2)

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Fit model with Stan.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

if model_name:

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    ### Model fitting and diagnostics.
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
        
    ## Fit model.
    file = 'stan_models/%s' %model_name
    fit = pystan.stan(file=file, data=dd, iter=samples, warmup=warmup, thin=thin, chains=chains, 
                      control=dict(adapt_delta = 0.9), n_jobs=n_jobs, seed=47404)
    check_div(fit); check_treedepth(fit); check_energy(fit)

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    ### Save data.
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

    ## Create out-directory.
    out_dir = 'stan_fits/%s' %model_name.replace('.stan','')
    if not os.path.isdir(out_dir): os.makedirs(out_dir)

    ## Save summary file.
    summary = fit.summary()
    summary = DataFrame(summary['summary'], columns=summary['summary_colnames'], index=summary['summary_rownames'])
    summary.to_csv(os.path.join(out_dir, 'summary.csv'))

    ## Save contents of StanFit.
    extract = fit.extract()
    for k, v in dd.items(): extract[k] = v
    with open(os.path.join(out_dir, 'StanFit.pickle'), 'wb') as fn: cPickle.dump(extract, fn)
        
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    ### Plot subjects.
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

    ## Make plots dir.
    plots_dir = os.path.join(out_dir, 'plots')
    if not os.path.isdir(plots_dir): os.makedirs(plots_dir)

    ## Iteratively plot.
    for i in np.arange(extract['N']):

        ## Plot and save.
        fig, axes = plot_subject(extract, i)
        fig.savefig(os.path.join(plots_dir, 'subj_%s.png' %(i+1)), dpi=180)
        plt.close('all')
        
    print('Done.')
```

### Bias, learning, and RPEs

```python
from pandas import concat
from scripts.utilities import HDIofMCMC, zscore

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Compute reward prediction errors.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Define useful local function.
fit = load_fit('moodRL_centered_full_mood_glob')

## Preallocate space.
Q = np.zeros((2, fit['N'], 9))

## Main loop.
for ix in np.arange(2): 

    for i in np.arange(fit['N']):
        
        ## Extract subject task.
        x, y, r = [fit[param][i].copy() for param in ['X','Y','R']]
        x -= 1
        y -= 1
        
        ## Extract subject parameters.
        eta_v = np.median(fit['eta_v'], axis=0)[i]
        if not ix: 
            eta_h = 0
            f = 1
        else: 
            eta_h = np.median(fit['eta_h'], axis=0)[i]
            f = np.median(fit['f'], axis=0)[i]
            
        beta_h = np.median(fit['beta_h'], axis=0)[i]
        m2 = fit['m2']
        
        ## Initialize values.
        h = 0
        m = np.tanh(beta_h)
        
        for j in np.arange(fit['B']):
            
            ## If second block, update reward history/mood 
            ## to reflect Wheel of Fortune.
            if j == 1:
                m = m2[i]
                h = np.arctanh(m)
            
            for k in np.arange(fit['T']):
                                
                ## Compute reward prediction error.
                delta = (f ** m) * r[j,k] - Q[ix, i, x[j,k,y[j,k]]]
                
                ## Update expectations.
                Q[ix, i, x[j,k,y[j,k]]] += eta_v * delta
                
                ## Update history of rewards.
                h += eta_h * (delta - h)
                
                ## Update mood.
                m = np.tanh( beta_h + h )

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Compute block differences.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Reshape.
Q = Q.reshape(2,fit['N'],3,3)

## Compute difference across blocks.
b12 = (Q[:,:,1] - Q[:,:,0]).reshape(2, fit['N'] * 3)
b13 = (Q[:,:,2] - Q[:,:,0]).reshape(2, fit['N'] * 3)

## Assemble into DataFrame.
b12 = DataFrame(np.vstack([np.repeat(np.arange(fit['N']), 3).astype(int)+1, 
                           np.zeros(b12.shape[-1]),  
                           np.concatenate([[20,40,60] for _ in np.arange(fit['N'])]).astype(int),
                           *b12]).T, 
                columns=['Subject','Block','Machine','No Mood','Full Mood'])

b13 = DataFrame(np.vstack([np.repeat(np.arange(fit['N']), 3).astype(int)+1, 
                           np.ones(b13.shape[-1]),  
                           np.concatenate([[20,40,60] for _ in np.arange(fit['N'])]).astype(int),
                           *b13]).T, 
                columns=['Subject','Block','Machine','No Mood','Full Mood'])

Q = concat([b12,b13])

## Merge with block preference data.
df = read_csv('data/moodRL_preferences.csv')
df = df.merge(Q, on=['Subject','Block','Machine'])

## Compute difference between model predictions.
df['delta'] = df['Full Mood'] - df['No Mood']

## Z-score.
for col in ['Likelihood', 'No Mood', 'Full Mood', 'delta']: df[col] = zscore(df[col])

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Fit model with Stan.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Assemble design matrix for regression.
X = np.vstack([
    
        ## Main effects.
        np.ones(df.shape[0]),
        df['No Mood'].values,
        df.delta.values
    
    ]).T.astype(float)

## Assemble data dictionary.
dd = dict(N = X.shape[0], M = X.shape[1], Y = df.postWoF.values, T = df.Total.values, X = X)

## Fit model.
file = 'stan_models/moodRL_block_preference.stan'
fit = pystan.stan(file=file, data=dd, iter=2500, warmup=2000, thin=1, chains=4, 
                  n_jobs=4, seed=47404)
check_div(fit); check_treedepth(fit); check_energy(fit)

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Plotting.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Extract parameters.
extract = fit.extract()
beta = extract['beta']

## Initialize canvas.
fig, axes = plt.subplots(1, 3, figsize=(12,3), sharey=True)

## Plot.
labels = ['Intercept', '$Post - Pre$\nEV (No Mood)', '$Post - Pre$\ndEV (Full Mood)']

for i, ax, arr, label in zip(range(3), axes, beta.T, labels):
    
    ## Select color.
    lb, ub = HDIofMCMC(arr, 0.95)
    if np.logical_xor(lb > 0, ub > 0): color='grey'
    else: color = '#52315f'
    
    ## Plot.
    sns.distplot(arr, bins=20, kde=False, color=color, hist_kws=dict(edgecolor='w'), ax=ax)
    
    ## Plot HDI.
    ax.vlines([lb,ub],0,300,lw=0.5,linestyle='--')
    ax.vlines(0,0,10,lw=2,color='k')
    ax.set(xlim=(-0.75,0.75), ylim=(0,300), xlabel=label)

axes[0].set(yticks=[0,150,300], ylabel='Count')
sns.despine()
plt.tight_layout()
```

### Simulate data

```python
from scripts.simulations import MoodyAgent, slot_machine_game

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Define parameters.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Simulation parameters.
Ns = 56    # subjects
B = 3      # blocks
T = 42     # trials

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Simulate data.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
np.random.seed(47404)

## Simulate mood.
mu = np.random.uniform(-2,2,Ns)
Ms = np.random.multivariate_normal(mu, np.identity(Ns)*0.7, B*3+1).T / 4
Ms = np.where(Ms < -1, -0.99, np.where(Ms > 1, 0.99, Ms))
m2s = Ms[:,-1]
Ms = Ms[:,:-1]

## Simulate agent parameters.
mu = [9.5, 0.09]
cov = [[1, 3e-3],
       [3e-3, 1e-4]]
beta, eta_v = np.random.multivariate_normal(mu, cov, Ns).T

## Preallocate space.
Xs = np.zeros([Ns,B,T,2], dtype=int)
Ys = np.zeros([Ns,B,T], dtype=int)
Rs = np.zeros([Ns,B,T], dtype=int)

## Simulate behavior.
for i in np.arange(Ns):
    
    ## Initialize agent.
    agent = MoodyAgent(beta=beta[i], eta_v=eta_v[i], eta_h=0, f=1)
    
    for j in np.arange(B):
        
        ## Simulate block of trials.
        x, r = slot_machine_game(n_trials=T, probabilities=(0.2,0.4,0.6), reward=1)
        
        ## Simulate behavior.
        y, _ = agent.simulate(x,r)
        
        ## Store.
        Xs[i,j] = x.copy() + j * B
        Ys[i,j] = y.copy()
        Rs[i,j] = [r[a,b] for a,b in enumerate(y)]

## Make arrays 1-indexed.
Xs += 1
Ys += 1
     
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Plotting.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Initialize canvas.
fig, axes = plt.subplots(1,4,figsize=(12,4))
sns.set_context('notebook', font_scale=1.75)

## Compare beta.
axes[0].scatter(np.sort(np.median(fit['beta'],axis=0)), np.sort(beta), color='#c17e9a')
axes[0].plot([5,12.5],[5,12.5], color='k')
axes[0].set(xlim=(5,12.5),ylim=(5,12.5),xlabel='Observed',
            ylabel='Simulated', title=r'$\beta$')

## Compare eta_v.
axes[1].scatter(np.sort(np.median(fit['eta_v'],axis=0)), np.sort(eta_v), color='#c17e9a')
axes[1].plot([0.06,0.12],[0.06,0.12], color='k')
axes[1].set(xlim=(0.06,0.12),ylim=(0.06,0.12),xlabel='Observed',
            ylabel='Simulated', yticks=np.arange(0.06,0.13,0.02),title=r'$\eta_v$')

## Compare mood mean.
axes[2].scatter(np.sort(M.mean(axis=1)), np.sort(Ms.mean(axis=1)), color='#c17e9a')
axes[2].plot([-0.5,0.5],[-0.5,0.5], color='k')
axes[2].set(xlim=(-0.5,0.5), xticks=np.arange(-0.4,0.5,0.2), ylim=(-0.5,0.5), xlabel='Observed',
            ylabel='Simulated', yticks=np.arange(-0.4,0.5,0.2), title='$M_\mu$')

## Compare mood std.
axes[3].scatter(np.sort(M.std(axis=1)), np.sort(Ms.std(axis=1)), color='#c17e9a')
axes[3].plot([0.0,0.3],[0.0,0.3], color='k')
axes[3].set(xlim=(0.0,0.3),ylim=(0.0,0.3),yticks=np.arange(0,0.4,0.1), xlabel='Observed',
            ylabel='Simulated',title='$M_\sigma$')

sns.despine()
plt.tight_layout()
```

### Model fitting

```python
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Define parameters.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Select model.
model_name = False

## Sampling parameters.
samples = 1250
warmup = 1000
chains = 4
thin = 1
n_jobs = 4

#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
### Fit model with Stan.
#~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

## Organize data dictionary.

dd = dict(N=Ns, B=B, T=T, X=Xs, Y=Ys, R=Rs, M=Ms.reshape(Ns, B, 3), m2=m2s)

if model_name:

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    ### Model fitting and diagnostics.
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
        
    ## Fit model.
    file = 'stan_models/%s' %model_name
    fit = pystan.stan(file=file, data=dd, iter=samples, warmup=warmup, thin=thin, chains=chains, 
                      control=dict(adapt_delta = 0.9), n_jobs=n_jobs, seed=47404)
    check_div(fit); check_treedepth(fit); check_energy(fit)

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    ### Save data.
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

    ## Create out-directory.
    out_dir = 'stan_fits/%s_recovery' %model_name.replace('.stan','')
    if not os.path.isdir(out_dir): os.makedirs(out_dir)

    ## Save summary file.
    summary = fit.summary()
    summary = DataFrame(summary['summary'], columns=summary['summary_colnames'], index=summary['summary_rownames'])
    summary.to_csv(os.path.join(out_dir, 'summary.csv'))

    ## Save contents of StanFit.
    extract = fit.extract()
    for k, v in dd.items(): extract[k] = v
    with open(os.path.join(out_dir, 'StanFit.pickle'), 'wb') as fn: cPickle.dump(extract, fn)

    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#
    ### Plot subjects.
    #~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~~#

    ## Make plots dir.
    plots_dir = os.path.join(out_dir, 'plots')
    if not os.path.isdir(plots_dir): os.makedirs(plots_dir)

    ## Iteratively plot.
    for i in np.arange(extract['N']):

        ## Plot and save.
        fig, axes = plot_subject(extract, i)
        fig.savefig(os.path.join(plots_dir, 'subj_%s.png' %(i+1)), dpi=180)
        plt.close('all')

print('Done.')
```

```python
from pandas import concat

models = ['moodRL_centered_full_mood_glob', 'moodRL_centered_full_mood_glob_recovery']
labels = ['Observed', 'Simulated']

df = []
for model, label in zip(models, labels):
    
    ## Load StanFit.
    fit = load_fit(model)
    
    ## Prepare mood data.
    M = fit['M']
    h_pred = ( np.median(fit['h_pred'], axis=0).T + np.median(fit['beta_h'], axis=0) ).T
    Mhat = np.tanh(h_pred[...,[7,21,35]])
    
    ## Compute correlations.
    rvals = np.array([pearsonr(x.flatten(),y.flatten())[0] for x,y in zip(M,Mhat)])
    rvals = rvals[~np.isnan(rvals)]
    df.append(DataFrame(np.array([rvals, np.repeat(label,rvals.size)]).T,
              columns = ('R', 'Model')))
    
    ## Perform t-test.
    t, p = ttest_1samp(rvals, 0)
    print('%s: t=%0.2f, p=%0.3f' %(label,t,p))
    
df = concat(df)

df['R'] = df.R.astype(float)

## Initialize canvas.
fig, ax = plt.subplots(1,1,figsize=(10,4))
sns.stripplot('Model', 'R', data=df,  palette=['#52315f', 'grey'], jitter=True, zorder=0, ax=ax)
for i, model in enumerate(df.Model.unique()):
    ax.hlines(df.loc[df.Model==model,'R'].mean(), i-0.175, i+0.175)
ax.hlines(0,-0.5,2.5,lw=0.5,linestyle='--',zorder=0)
ax.set(xlabel='', ylabel=r'Pearsonr $r$')

sns.despine()
plt.tight_layout()
```
